```python
class DataCrawling(object):
    
    """This function takes in a list of urls and return the html_list
    Input: a list of url
    Output: a tuple of html list and inaccessible url list"""
    
    def __call__(self, file_name):
        self.url = pd.read_excel(os.path.join('./data',file_name))['url']

        html = []
        id_list = []

        for i, url in enumerate(list(self.url)):
            try:
                page = requests.get(url, timeout=10) 
                content = page.text
                html.append(content)
                id_list.append(i)
                print('finish step '+str(i))
            except:
                html.append('None')
                id_list.append(i)
                print('Exception occur when retrieve:', url)

        return (html, id_list)
 
```

```python
def read_train_file(file_name):
    mylist = []

    for chunk in  pd.read_csv(os.path.join('./data',file_name), header=None, chunksize=20):
        mylist.append(chunk)

    big_data = pd.concat(mylist, axis= 0).iloc[:,1]
    del mylist
    return big_data
```

```python
def preprocess_text(lists):
    preprocess_text_list=[]
    for i in range(len(lists)):
        if lists[i]=='None':
            preprocess_text_list.append('None')
            print('None  ', i)
        else:
            try:
                preprocess_text_list.append(text_process.text_processing(lists[i]))
                print('success  ', i)
            except:
                preprocess_text_list.append('None')
    return (preprocess_text_list)
```

```python
class MeanEmbeddingVectorizer(object):
    def __init__(self, wordtwovec):
        self.wordtwovec = wordtwovec
        # if a text is empty we should return a vector of zeros
        # with the same dimensionality as all the other vectors
        self.dim = len(next(iter(wordtwovec.values())))

    def fit(self, X, y):
        return self

    def transform(self, X):
        return np.array([
            np.mean([self.wordtwovec[w] for w in words if w in self.wordtwovec]
                    or [np.zeros(self.dim)], axis=0)
            for words in X
        ])
```

```python
def Word2Vec_Processing(train_set,test_set):
    # let X be a list of tokenized texts (i.e. list of lists of tokens)
    train_w2v=Word2Vec(train_set, size=100)
    test_w2v=Word2Vec(test_set, size=100)
    train_w2v_dict = dict(zip(train_w2v.wv.index2word, train_w2v.wv.syn0))
    test_w2v_dict = dict(zip(test_w2v.wv.index2word, test_w2v.wv.syn0))
    # get vector data
    train_vector = MeanEmbeddingVectorizer(train_w2v_dict).transform(train_set)
    test_vector=MeanEmbeddingVectorizer(test_w2v_dict).transform(test_set)
    return train_vector, test_vector
```

```python
def clustering(X, Y, num_cluster):
    NUM_CLUSTERS = num_cluster
    kmeans = cluster.KMeans(n_clusters=NUM_CLUSTERS)
    kmeans.fit(X)
    prediction=kmeans.predict(Y)
    return (prediction)
```

```python
def result_extracting(prediction):
    df=pd.read_excel(os.path.join('./data','test.xlsx'))
    df['prediction']=prediction
    df.to_excel(os.path.join('./output','final_result.xlsx'))
```

```python
#main
def main():
    crawling=DataCrawling()
    test_set , id_list=crawling('test.xlsx')
    train_set=read_train_file('train3.csv')
    test_set= preprocess_text(test_set)
    train_set = preprocess_text(train_set)
    X, Y = Word2Vec_Processing(train_set,test_set)
    prediction = clustering(X, Y, 2)
    result_extracting(prediction)
 
```
